```python
#массив предложений(строк)
import re
with open('/Users/volchkov/Downloads/sentences.txt', 'r') as f:
  sentences = []
  for line in f:
    line_list = re.split('[^a-z]', line.strip('\\n\\r.').lower())
    line_list = list(filter(None, line_list))        
    sentences.append(line_list)
```

```python
#Создайте матрицу размера n * d, где n — число предложений. 
#Заполните ее: элемент с индексом (i, j) в этой матрице должен быть равен количеству вхождений j-го слова в i-е предложение. 
#У вас должна получиться матрица размера 22 * 254.
import numpy as np
n = len(open('/Users/volchkov/Downloads/sentences.txt', 'r').readlines()) #число предложений
d = len(dict_a)
word_matrix = np.zeros(shape=(n,d))
for i in range(n):
    for j in range(d):
        word_matrix[i][j]=sentences[i].count(dict_a[j])
```

```python
import math
X=np.array([[1, 1], [1, 15]])
Y=np.array([[math.sin(1/5)*math.exp(1/10)+5*math.exp(-1/2)], [math.sin(15/5)*math.exp(15/10)+5*math.exp(-15/2)]])
```

```python
Y
```

```python
solving = np.linalg.solve(X, Y)
```

```python
solving[1]
```

```python
#Импортируем библиотеку Math
import math
#Импортируем один из пакетов Matplotlib
import pylab
#Импортируем пакет со вспомогательными функциями
from matplotlib import mlab


def func (x):
    return solving[0]+solving[1]*x      
    
#Указываем X наименьее и наибольшее
xmin = -8.0
xmax = 8.0

# Шаг между точками
dx = 0.01

#Создадим список координат по оси 
#X на отрезке [-xmin; xmax], включая концы
xlist = mlab.frange (xmin, xmax, dx)

# Вычислим значение функции в заданных точках
ylist = [func (x) for x in xlist]

#Нарисуем одномерный график
pylab.plot (xlist, ylist)

#Покажем окно с нарисованным графиком
pylab.show()
```
